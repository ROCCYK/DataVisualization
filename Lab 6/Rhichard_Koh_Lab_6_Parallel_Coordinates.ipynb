```python
from ucimlrepo import fetch_ucirepo

# fetch dataset
wine_quality = fetch_ucirepo(id=186)
wine_quality = wine_quality.data.original

three = wine_quality.loc[wine_quality['quality'] == 3]
nine = wine_quality.loc[wine_quality['quality'] == 9]
both = wine_quality.loc[(wine_quality['quality'] == 9) | (wine_quality['quality'] == 3)]
```

```python
import plotly.express as px
color_discrete_map = ['blue','red','black','yellow','purple','orange','pink']
fig = px.parallel_coordinates(wine_quality, color='quality', color_continuous_scale=color_discrete_map)
fig.show()
```

```python
import plotly.express as px
fig = px.parallel_coordinates(three, color='quality',color_continuous_scale=color_discrete_map)
fig.show()
```

```python
import plotly.express as px
fig = px.parallel_coordinates(nine, color='quality',color_continuous_scale=color_discrete_map)
fig.show()
```

```python
import plotly.express as px
fig = px.parallel_coordinates(both, color='quality',color_continuous_scale=color_discrete_map)
fig.show()
```
